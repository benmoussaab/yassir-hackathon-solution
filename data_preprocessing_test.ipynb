```python
import pandas as pd
import numpy as np
from datetime import datetime
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('data/test_normelized_v2.csv', low_memory=False)
df.head()
```

```python
len(df)
```

```python
df.drop(columns=['pickup','CommissionPercent','epay_method', 'epay_type', 'destination'], inplace=True)
```

```python
df.head()
```

```python
# Convert date-time columns to datetime
date_columns = ['requested_at', 'updated_at', 'created_at', 'accepted_at', 
                'started_at', 'cancel_date_time']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='mixed')  # or omit format if mixed date formats are present
df.head()
```

```python
df['request_hour'] = df['requested_at'].dt.hour
df['request_day_of_week'] = df['requested_at'].dt.dayofweek
df['request_month'] = df['requested_at'].dt.month
df['request_day_of_month'] = df['requested_at'].dt.day
df.head()
```

```python
df['drivers_list'] = df['drivers_list'].apply(lambda x: x.split(',') if x is not None else [])
```

```python
df['drivers_list'].iloc[0]
```

```python
df = df.explode('drivers_list')
```

```python
df['drivers_list'] = df['drivers_list'].str.strip()
```

```python
df['drivers_list'].iloc[1]
```

```python
df.head()
```

```python
df['drivers_list'].iloc[1]
```

```python
df['driver'] = df['drivers_list']
df.drop(columns=['drivers_list'], inplace=True)
df.head()
```

```python
df.to_csv('data/test_data_v2_exploded.csv', index=False)
```

```python
len(df)
```

```python
acceptance_rate_pivot = pd.read_csv('data/acceptance_rate_pivot.csv')
```

```python
acceptance_rate_pivot.head()
```

```python
df = df.merge(acceptance_rate_pivot, on='driver', how='left')
df.head()
```

```python
df.info()
```

```python
columns_to_fill = ['acceptance_rate_monday', 'acceptance_rate_tuesday', 'acceptance_rate_wednesday', 'acceptance_rate_thursday',
                   'acceptance_rate_friday', 'acceptance_rate_saturday', 'acceptance_rate_sunday'
                   ]
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].median())
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
driver_acceptance_stats = pd.read_csv('data/driver_acceptance_stats.csv')
```

```python

# Step 3: Merge these stats back into the original dataframe
df = df.merge(driver_acceptance_stats[['driver', 'acceptance_rate', 'total_accepted']], 
              on='driver', how='left')

# Rename for clarity in the final dataframe
df.rename(columns={'total_accepted': 'acceptance_count'}, inplace=True)
```

```python
df.info()
```

```python
columns_to_fill = ['acceptance_rate', 'acceptance_count']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(columns=['requested_at', 'updated_at', 'created_at', 'accepted_at', 'started_at',
                 'cancel_date_time', 'rider_profile_gender', 'is_coupon',
                 'isMultiStop', 'is_booked', 'radius', 'dcc_distance',
                  'rider_rating', 
                 'discount_ratio_percent', 'request_day_of_month',
                 'rider', 'driver', 'trip_id', 'final_status'
                 ], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
right_order =   ['original_cost', 'rider_profile_rating', 'estimated_distance',
       'estimated_eta', 'discount', 'yassir2driver',
       'driver2yassir', 'request_hour', 'request_day_of_week',
       'request_month', 'acceptance_rate_monday', 'acceptance_rate_tuesday',
       'acceptance_rate_wednesday', 'acceptance_rate_thursday',
       'acceptance_rate_friday', 'acceptance_rate_saturday',
       'acceptance_rate_sunday', 'acceptance_rate', 'acceptance_count'
       ]
df = df[right_order]
```
