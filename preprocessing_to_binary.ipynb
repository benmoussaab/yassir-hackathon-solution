```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('/Users/hodhaifabenouaklil/Documents/Me/Competitions/YassirHackathon/V4/data/data_normelized_v2.csv', low_memory=False)
train.head()
```

```python
train['assigned_at'] = pd.to_datetime(train['assigned_at'], format='mixed')
```

```python
train =train.sort_values(by=['trip_id', 'assigned_at'])
train.head()
```

```python
train['assigined_driver'] = train['assigined_driver'].apply(lambda x: x.replace(" ", ""))
```

```python
train['final_driver'] = train['final_driver'].apply(lambda x: x.replace(" ", ""))
```

```python
train['final_driver'].str.contains(" ").unique()
```

```python
train['assigined_driver'].str.contains(" ").unique()
```

```python
def assign_driver_and_took_trip(group):
    # Initialize 'driver' and 'took_trip' with default values
    group['driver'] = group['assigined_driver']
    group['took_trip'] = 0
    
    # Duplicate the last row of the group
    last_row = group.iloc[[-1]]  # Extract the last row
    group = pd.concat([group, last_row], ignore_index=True)  # Append the last row to the group
    
    # Modify the duplicated last row
    group.loc[group.index[-1], 'driver'] = group.loc[group.index[-1], 'final_driver']
    group.loc[group.index[-1], 'took_trip'] = 1
    
    return group

# Apply the function to each group
train = train.sort_values(by=['trip_id', 'assigned_at'])  # Ensure proper order
train = train.groupby('trip_id', group_keys=False).apply(assign_driver_and_took_trip)
```

```python
train.head()
```

```python
train.drop(columns=['assigined_driver', 'final_driver'], inplace=True)
```

```python
train.to_csv('/Users/hodhaifabenouaklil/Documents/Me/Competitions/YassirHackathon/V4/data/data_normelized_binary_v2.csv', index=False)
```
