```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('data/data_normelized_binary_v2.csv', low_memory=False)
```

```python
df.drop(columns=['trip_id', 'pickup','CommissionPercent','epay_method', 'epay_type',
                 'assigned_at_rejection_type', 'destination', 'first_finished_driver'], inplace=True)
```

```python

# Convert date-time columns to datetime
date_columns = ['requested_at', 'updated_at', 'created_at', 'accepted_at', 
                'started_at', 'cancel_date_time', 'assigned_at']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='mixed')  # or omit format if mixed date formats are present
df.head()
```

```python
df['request_hour'] = df['requested_at'].dt.hour
df['request_day_of_week'] = df['requested_at'].dt.dayofweek
df['request_month'] = df['requested_at'].dt.month
df['request_day_of_month'] = df['requested_at'].dt.day
df.head()
```

```python
df.columns
```

```python
acceptance_rate = df.groupby(['driver', 'request_day_of_week'])['took_trip'].agg(
    total_requests=('count'),
    accepted_trips=('sum'),
    acceptance_rate=('mean')
).reset_index()

# Rename columns for clarity
acceptance_rate.columns = [
    'Driver', 
    'Day of Week', 
    'Total Requests', 
    'Accepted Trips', 
    'Acceptance Rate'
]

# Sort values for readability
acceptance_rate.sort_values(['Driver', 'Day of Week'], inplace=True)
acceptance_rate.head()
```

```python
acceptance_rates = df.groupby(['driver', 'request_day_of_week'])['took_trip'].mean().reset_index()
acceptance_rates.head()
```

```python
acceptance_rates.rename(columns={'took_trip': 'average_acceptance_rate'}, inplace=True)

# Step 3: Pivot the table to create 7 columns, one for each day of the week
acceptance_rate_pivot = acceptance_rates.pivot(index='driver', 
                                               columns='request_day_of_week', 
                                               values='average_acceptance_rate')

# Rename columns for clarity
acceptance_rate_pivot.columns = [
    'acceptance_rate_monday', 'acceptance_rate_tuesday', 'acceptance_rate_wednesday',
    'acceptance_rate_thursday', 'acceptance_rate_friday', 'acceptance_rate_saturday',
    'acceptance_rate_sunday'
]

# Step 4: Merge the pivoted table back into the original dataframe based on the 'driver' column
df = df.merge(acceptance_rate_pivot, on='driver', how='left')
df.head()
```

```python
acceptance_rate_pivot.fillna(0., inplace=True)
acceptance_rate_pivot.head()
```

```python
acceptance_rate_pivot.to_csv('data/acceptance_rate_pivot.csv')
```

```python
driver_acceptance_stats = df.groupby('driver')['took_trip'].agg(
    total_requests='count',  # Total number of requests per driver
    total_accepted='sum'    # Total number of accepted requests per driver
).reset_index()

# Step 2: Calculate the acceptance rate
driver_acceptance_stats['acceptance_rate'] = (
    driver_acceptance_stats['total_accepted'] / driver_acceptance_stats['total_requests']
)

# Step 3: Merge these stats back into the original dataframe
df = df.merge(driver_acceptance_stats[['driver', 'acceptance_rate', 'total_accepted']], 
              on='driver', 
              how='left')

# Rename for clarity in the final dataframe
df.rename(columns={'total_accepted': 'acceptance_count'}, inplace=True)
```
