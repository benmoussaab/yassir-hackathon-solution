```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
X = pd.read_csv('data/data_final_test.csv')
X.head()
```

```python
X.drop(columns=['estimated_distance'], inplace=True)
```

```python
X.head()
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.load_model('xgboost_model.json')
```

```python
y_pred = model.predict(X)
y_pred_proba = model.predict_proba(X)[:, 1]
y_pred, y_pred_proba
```

```python
probabilities = model.predict_proba(X)[:, 1]  # Get probabilities for the positive class
```

```python
test = pd.read_csv('data/test_data_v2_exploded.csv')
test.head()
```

```python
len(X), len(test)
```

```python
test['probability'] = probabilities
```

```python
test.head()
```

```python
# test = test.sort_values(by=['probability'], ascending=[False])
test = test.sort_values(by=['trip_id', 'probability'], ascending=[True, False])
```

```python
test.head()
```

```python
test.iloc[0]['trip_id']
```

```python
test.iloc[0]['driver']
```

```python
test.iloc[1]['driver']
```

```python
result = (
    test.sort_values(by=['trip_id', 'probability'], ascending=[True, False])  # Ensure correct order
    .groupby('trip_id')
    .agg({'driver': list})  # Aggregate driver_id into a list
    .reset_index()  # Optional: reset the index for a clean DataFrame
)

# Rename the aggregated column if needed
result.rename(columns={'driver': 'drivers_list'}, inplace=True)

result.head()
```

```python
result['drivers_ranking'] = result['drivers_list']
result.drop(columns=['drivers_list'], inplace=True)
```

```python
result.to_csv('data/submission.csv', index=False)
```
